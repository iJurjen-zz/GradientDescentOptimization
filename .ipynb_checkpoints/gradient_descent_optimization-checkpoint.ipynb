# Gradient Descent Optimization

```python
import numpy as np
from cvxpy import *
import matplotlib.pyplot as plt
from sklearn import preprocessing

% matplotlib inline
```

First I'll use the cvxpy library to find the least square solution of the best linear regression fit to a small 2D dataset. I will use this data later to check the stochastic gradient descent algorithm.

```python
data = np.loadtxt('./PierceCricketData.csv', delimiter=',')
(m,n) = data.shape
A = np.append(data[:,0].reshape(m,1), np.ones((m,1)), axis=1)
temp = data[:,1].reshape(m,1)

x = Variable(n)
objective = Minimize(sum_squares(A*x - temp))
prob = Problem(objective)

# Imprimimos el objetivo optimo
result = prob.solve()
print(result)
# Imprimimos el valor optimo de x
print(x.value)
```

```python
#Representamos el resultado
(theta_0, theta_1) = (float(x.value[1]), float(x.value[0]))
print(theta_0, theta_1)
f = lambda theta_0,theta_1,x: theta_0 + theta_1*x
xx = np.linspace(0,21,1000)
plt.scatter(data[:, 0], data[:, 1], marker='o', c='b')
plt.plot(xx,f(theta_0, theta_1,xx))
plt.xlim([13,21])
plt.ylim([65,95])
plt.title('cricket chirps vs temperature')
plt.xlabel('chirps/sec for striped ground crickets')
plt.ylabel('temperature in degrees Fahrenheit')
plt.show()
```

# Stochastic Gradient Descent

```python
def descenso_estocastico_gradiente(X, y, theta_ini, pasadas, eta_ini, eta_adjust=False, plot_cost=False):
    """
    Implementa el algoritmo de descenso estocástico de gradiente
    para regresión lineal por mínimos cuadrados.
    
    Parámetros:
    x         variable independiente de la regresión lineal. Array de dimensiones mxn. Primer columna de unos.
    y         variable dependiente de la regresión lineal. Array unidimensional de m elementos
    theta_ini valor inicial de los parámetros a optimizar. Array unidimensional de n elementos.
    pasadas   número de veces que recorreremos cada uno de los m elementos del conjunto de datos
    eta       longitud de paso
    
    Valor devuelto:
    Array con los valores óptimos de theta. Array unidimensional de n elementos.
    """
    
    #TODO: implementar esta función
    
    (m,n) = X.shape
    data = np.concatenate((X, y.reshape((m,1))), axis=1)   # Necesary for the data shuffle
    theta = np.copy(theta_ini).reshape(n,1)
    eta = np.copy(eta_ini)
    Costs = []

    for pasada in range(pasadas):
        
        # Randomly shuffle dataset
        np.random.shuffle(data)
        X = data[:, :-1]
        y = data[:, -1].reshape(m,1)
    
        # Loop through the m examples
        for i in range(m):
            x_i = X[i,:].reshape(1,n)
            y_i = y[i]
            y_hat_i = np.dot(x_i, theta)   # estimation in single point
            error_i = y_hat_i - y_i   # error in this point
            grad = error_i * x_i   # gradient at this point
            theta -= eta * grad.reshape(n,1)   # update theta
        
        if eta_adjust == True:
            eta = eta_ini / (1+pasada)   # adjust learning rate
        
        y_hat = np.matmul(X, theta)
        error = y_hat.reshape(m,1) - y
        cost = 0.5 * np.sum(error**2) / m
        Costs.append(cost)
    
    if plot_cost == True:
        xx = np.linspace(1, pasadas, pasadas)
        plt.plot(xx, Costs, c='b')
        #plt.ylim(0, 200, auto=True)
        plt.title("Stochastic GradientDescent")
        plt.xlabel('pasada')
        plt.ylabel('cost')
    
    return theta
```

First I'll check the algorithm on the PierceCricket data:

```python
data = np.loadtxt('./PierceCricketData.csv', delimiter=',')
(m,n) = data.shape
X = np.concatenate((np.ones((m,1)).reshape(m,1), data[:,0].reshape(m,1)), axis=1)
temp = data[:,1].reshape(m,1)

theta_ini = np.random.rand(n)
theta = descenso_estocastico_gradiente(X, temp, theta_ini, 10**5, 0.001, plot_cost=True)
print(theta)
```

The value of eta needs some tuning, but with this learning rate it is getting close after 100.000 epochs.  
To compare with the solution I got with cvxpy:

```python
y_hat = np.matmul(X, theta)
error = y_hat - temp
cost = np.sum(error**2)
print("Cost with stochastic gradient descent: ", round(cost,2))

theta_cvxpy = np.array([26.012043411569508, 3.244165727621934]).reshape(2,1)
y_hat = np.matmul(X, theta_cvxpy)
error = y_hat - temp
cost = np.sum(error**2)
print("cost_min = ", round(cost,2))
```

```python
f = lambda theta_0,theta_1,x: theta_0 + theta_1*x
xx = np.linspace(0,21,1000)
plt.scatter(data[:, 0], data[:, 1], marker='o', c='b')
plt.plot(xx,f(theta[0], theta[1], xx), c='r')
plt.plot(xx,f(theta_cvxpy[0], theta_cvxpy[1], xx), c='b')
plt.xlim([13,21])
plt.ylim([65,95])
plt.title('solution with stochastic gradient descent in red')
plt.xlabel('chirps/sec for striped ground crickets')
plt.ylabel('temperature in degrees Fahrenheit')
plt.show()
```

 ### Let's implement mini-batch gradient descent:

```python
def mini_batch_gradientdescent(X, y, theta_ini, pasadas, eta_ini, b=16):
    """
    Implementa el algoritmo de descenso estocástico de gradiente
    para regresión lineal por mínimos cuadrados.
    
    Parámetros:
    x         variable independiente de la regresión lineal. Array de dimensiones mxn. Primer columna de unos.
    y         variable dependiente de la regresión lineal. Array unidimensional de m elementos
    theta_ini valor inicial de los parámetros a optimizar. Array unidimensional de n elementos.
    pasadas   número de veces que recorreremos cada uno de los m elementos del conjunto de datos
    eta       longitud de paso
    b         mini-batch size
    
    Valor devuelto:
    Array con los valores óptimos de theta. Array unidimensional de n elementos.
    """
    
    #TODO: implementar esta función
    
    (m,n) = X.shape
    data = np.concatenate((X, y.reshape((m,1))), axis=1)   # Necesary for the data shuffle
    theta = np.copy(theta_ini).reshape(n,1)
    Theta = np.copy(theta_ini).reshape(n,1)
    eta = np.copy(eta_ini)
    Costs = []

    for pasada in range(pasadas):
        
        # Randomly shuffle dataset
        np.random.shuffle(data)
        X = data[:, :-1]
        y = data[:, -1].reshape(m,1)
    
        # Loop through the m examples in batches of size b
        n_batch = np.floor(m/b).astype(int)
        for i in range(n_batch):
            start = b*i
            stop = b*i + b
            X_batch_i = X[start:stop, :]
            y_batch_i = y[start:stop]
            y_hat_batch_i = np.matmul(X_batch_i, theta)   # estimation in batch
            error_batch_i = y_hat_batch_i - y_batch_i   # error
            grad = np.sum((X_batch_i * error_batch_i.reshape(b,1)), axis=0) / b   # gradient
            theta -= eta * grad.reshape(n,1)   # update theta
                
        Theta = np.hstack((Theta, theta.reshape(n,1)))
        y_hat = np.matmul(X, theta)
        error = y_hat - y
        cost = 0.5 * np.sum(error**2) / m
        Costs.append(cost)

    plt.figure(1)
    xx1 = np.linspace(1, pasadas, pasadas)
    plt.plot(xx1, Costs, label = 'Cost', color='b')
    plt.ylim(0, 100)
    plt.legend()

    plt.figure(2)
    xx0 = np.linspace(0, pasadas, pasadas+1)
    for i in range(n):
        plt.plot(xx0, Theta[i,:], label = 'theta_%s'%i)
    plt.legend(loc=2)
    plt.title("Minibatch GradientDescent")
        
    return theta
```

```python
mini_batch_gradientdescent(x, y, theta_ini, 50, 0.0001)
```

The algoritm has a lot of trouble learning theta_0.  
This is because to caculate the gradient the error is multiplied by an x-value. For theta_0 this x-value is always 1, while for the other x-coordinates this x-value can vary between 0 and 100.  
### Normalizing X first:

```python
def mini_batch_gradientdescent(X, y, theta_ini, pasadas, eta_ini, b=16):
    """
    This version is WITH SCALING!!
    
    Implementa el algoritmo de descenso estocástico de gradiente
    para regresión lineal por mínimos cuadrados.
    
    Parámetros:
    x         variable independiente de la regresión lineal. Array de dimensiones mxn. Primer columna de unos.
    y         variable dependiente de la regresión lineal. Array unidimensional de m elementos
    theta_ini valor inicial de los parámetros a optimizar. Array unidimensional de n elementos.
    pasadas   número de veces que recorreremos cada uno de los m elementos del conjunto de datos
    eta       longitud de paso
    b         mini-batch size
    
    Valor devuelto:
    Array con los valores óptimos de theta. Array unidimensional de n elementos.
    """
    
    #TODO: implementar esta función
    
    (m,n) = X.shape
    data = np.concatenate((X, y.reshape((m,1))), axis=1)   # Necesary for the data shuffle
    theta = np.copy(theta_ini).reshape(n,1)
    Theta = np.copy(theta_ini).reshape(n,1)
    eta = np.copy(eta_ini)

    for pasada in range(pasadas):
        
        # Randomly shuffle dataset
        np.random.shuffle(data)
        X = data[:, :-1]
        y = data[:, -1].reshape(m,1)

        # Normalize X
        X_scaled = preprocessing.scale(X, axis=0, with_mean=False, with_std=True, copy=True)        
        
        # Loop through the m examples in batches of size b
        n_batch = np.floor(m/b).astype(int)
        for i in range(n_batch):
            start = b*i
            stop = b*i + b
            X_batch_i = X[start:stop, :]
            y_batch_i = y[start:stop]

            y_hat_batch_i = np.matmul(X_batch_i, theta)   # estimation in batch
            error_batch_i = y_hat_batch_i - y_batch_i   # error
            
            X_scaled_batch_i = X_scaled[start:stop, :]
            grad = np.sum((X_scaled_batch_i * error_batch_i.reshape(b,1)), axis=0) / b   # gradient
            theta -= eta * grad.reshape(n,1)   # update theta
        
        Theta = np.hstack((Theta, theta.reshape(n,1)))

    xx = np.linspace(0, pasadas, pasadas+1)
    for i in range(n):
        plt.plot(xx, Theta[i,:], label = 'theta_%s'%i)
    plt.legend(loc=1) 
    plt.title("Minibatch GradientDescent with normalizing")
    
    return theta
```

```python
mini_batch_gradientdescent(x, y, theta_ini, 20, 0.001)   # In the scaled version we can use a bigger value of eta
```

We can see that normalizing first is an improvement.
### Let's try to implement momentum:

```python
def mini_batch_gradientdescent(X, y, theta_ini, pasadas, eta_ini, beta=0.9, b=16):
    """
    This version is WITH MOMENTUM!!
    
    Implementa el algoritmo de descenso estocástico de gradiente
    para regresión lineal por mínimos cuadrados.
    
    Parámetros:
    x         variable independiente de la regresión lineal. Array de dimensiones mxn. Primer columna de unos.
    y         variable dependiente de la regresión lineal. Array unidimensional de m elementos
    theta_ini valor inicial de los parámetros a optimizar. Array unidimensional de n elementos.
    pasadas   número de veces que recorreremos cada uno de los m elementos del conjunto de datos
    eta       longitud de paso
    b         mini-batch size
    
    Valor devuelto:
    Array con los valores óptimos de theta. Array unidimensional de n elementos.
    """
    
    #TODO: implementar esta función
    
    (m,n) = X.shape
    data = np.concatenate((X, y.reshape((m,1))), axis=1)   # Necesary for the data shuffle
    theta = np.copy(theta_ini).reshape(n,1)
    Theta = np.copy(theta_ini).reshape(n,1)
    eta = np.copy(eta_ini)
    v_grad = np.zeros(n)   # this can be seen as the 'velocity'-term, where grad can be seen as the 'acceleration'

    for pasada in range(pasadas):
        
        # Randomly shuffle dataset
        np.random.shuffle(data)
        X = data[:, :-1]
        y = data[:, -1].reshape(m,1)
        
        # Normalize X
        X_scaled = preprocessing.scale(X, axis=0, with_mean=False, with_std=True, copy=True)
    
        # Loop through the m examples in batches of size b
        n_batch = np.floor(m/b).astype(int)
        for i in range(n_batch):
            start = b*i
            stop = b*i + b
            X_batch_i = X[start:stop, :]
            y_batch_i = y[start:stop]
            
            y_hat_batch_i = np.matmul(X_batch_i, theta)   # estimation in batch
            error_batch_i = y_hat_batch_i - y_batch_i   # error
            
            X_scaled_batch_i = X_scaled[start:stop, :]
            grad = np.sum((X_batch_i * error_batch_i.reshape(b,1)), axis=0) / b   # gradient
            v_grad = beta * v_grad + (1-beta) * grad   # calculating momentum
            theta -= eta * v_grad.reshape(n,1)   # update theta
                
        Theta = np.hstack((Theta, theta.reshape(n,1)))
        
    xx = np.linspace(0, pasadas, pasadas+1)
    for i in range(n):
        plt.plot(xx, Theta[i,:], label = 'theta_%s'%i)
    plt.legend(loc=1) 
    plt.title("Minibatch GradientDescent with momentum")
    
    return theta
```

```python
mini_batch_gradientdescent(x, y, theta_ini, 20, 0.001)
```

The effect of momentum doesn't seem that big. And without adjusting the learning rate the curves wiggle more.
### Finally, I will implement the Adam optimizer

```python
def mini_batch_gradientdescent(X, y, theta_ini, pasadas, eta_ini, beta1=0.9, beta2=0.99, eps=10**-8, b=16):
    """
    This version is WITH ADAM!!
    
    Implementa el algoritmo de descenso estocástico de gradiente
    para regresión lineal por mínimos cuadrados.
    
    Parámetros:
    x         variable independiente de la regresión lineal. Array de dimensiones mxn. Primer columna de unos.
    y         variable dependiente de la regresión lineal. Array unidimensional de m elementos
    theta_ini valor inicial de los parámetros a optimizar. Array unidimensional de n elementos.
    pasadas   número de veces que recorreremos cada uno de los m elementos del conjunto de datos
    eta       longitud de paso
    b         mini-batch size
    
    Valor devuelto:
    Array con los valores óptimos de theta. Array unidimensional de n elementos.
    """
    
    #TODO: implementar esta función
    
    (m,n) = X.shape
    data = np.concatenate((X, y.reshape((m,1))), axis=1)   # Necesary for the data shuffle
    theta = np.copy(theta_ini).reshape(n,1)
    Theta = np.copy(theta_ini).reshape(n,1)
    eta = np.copy(eta_ini)
    v_grad = np.zeros(n)
    s_grad = np.zeros(n)
    

    for pasada in range(pasadas):
        
        # Randomly shuffle dataset
        np.random.shuffle(data)
        X = data[:, :-1]
        y = data[:, -1].reshape(m,1)
        
        # Normalize X
        X_scaled = preprocessing.scale(X, axis=0, with_mean=False, with_std=True, copy=True)
    
        # Loop through the m examples in batches of size b
        n_batch = np.floor(m/b).astype(int)
        for i in range(n_batch):
            start = b*i
            stop = b*i + b
            X_batch_i = X[start:stop, :]
            y_batch_i = y[start:stop]
            
            y_hat_batch_i = np.matmul(X_batch_i, theta)   # estimation in batch
            error_batch_i = y_hat_batch_i - y_batch_i   # error
                        
            X_scaled_batch_i = X_scaled[start:stop, :]
            grad = np.sum((X_batch_i * error_batch_i.reshape(b,1)), axis=0) / b   # gradient
            v_grad = beta1 * v_grad + (1-beta1) * grad   # calculating momentum
            s_grad = beta2 * s_grad + (1-beta2) * grad**2   # calculating RMSprop
            v_corr = v_grad / (1-beta1**((pasada+1)*(i+1)))   # bias correction
            s_corr = s_grad / (1-beta2**((pasada+1)*(i+1)))   # bias correction
            theta -= eta * (v_corr / (np.sqrt(s_corr) + eps)).reshape(n,1)   # update theta
        
        # adjusting the learning rate; so we can start with a big value and decrease it every epoch
        eta = eta_ini / (1+10*pasada)
        
        Theta = np.hstack((Theta, theta))

    xx = np.linspace(0, pasadas, pasadas+1)
    for i in range(n):
        plt.plot(xx, Theta[i,:], label = 'theta_%s'%i)
    plt.legend(loc=1)
    plt.title("Minibatch GradientDescent with Adam optimizer and decreasing learning rate")
    plt.xlabel('pasada')
    
    return theta
```

```python
mini_batch_gradientdescent(x, y, theta_ini, 20, 1)   # With Adam we can use a bigger value of eta!
```

With Adam optimizer learning converges quickly. Let's look at what happens when we do 1 epoch over the 100.000 points.

We see that after 1 epoch it gets to pretty decent values of theta.

```python
def mini_batch_gradientdescent(X, y, theta_ini, pasadas, eta_ini, beta1=0.9, beta2=0.99, eps=10**-8, b=16):
    """
    This version is stores theta after every batch iteration.
    
    Implementa el algoritmo de descenso estocástico de gradiente
    para regresión lineal por mínimos cuadrados.
    
    Parámetros:
    x         variable independiente de la regresión lineal. Array de dimensiones mxn. Primer columna de unos.
    y         variable dependiente de la regresión lineal. Array unidimensional de m elementos
    theta_ini valor inicial de los parámetros a optimizar. Array unidimensional de n elementos.
    pasadas   número de veces que recorreremos cada uno de los m elementos del conjunto de datos
    eta       longitud de paso
    b         mini-batch size
    
    Valor devuelto:
    Array con los valores óptimos de theta. Array unidimensional de n elementos.
    """
    
    #TODO: implementar esta función
    
    (m,n) = X.shape
    data = np.concatenate((X, y.reshape((m,1))), axis=1)   # Necesary for the data shuffle
    theta = np.copy(theta_ini).reshape(n,1)
    Theta = np.copy(theta_ini).reshape(n,1)
    eta = np.copy(eta_ini)
    v_grad = np.zeros(n)
    s_grad = np.zeros(n)
    

    for pasada in range(pasadas):
        
        # Randomly shuffle dataset
        np.random.shuffle(data)
        X = data[:, :-1]
        y = data[:, -1].reshape(m,1)
        
        # Normalize X
        X_scaled = preprocessing.scale(X, axis=0, with_mean=False, with_std=True, copy=True)
    
        # Loop through the m examples in batches of size b
        n_batch = np.floor(m/b).astype(int)
        for i in range(n_batch):
            start = b*i
            stop = b*i + b
            X_batch_i = X[start:stop, :]
            y_batch_i = y[start:stop]
            
            y_hat_batch_i = np.matmul(X_batch_i, theta)   # estimation in batch
            error_batch_i = y_hat_batch_i - y_batch_i   # error
                        
            X_scaled_batch_i = X_scaled[start:stop, :]
            grad = np.sum((X_batch_i * error_batch_i.reshape(b,1)), axis=0) / b   # gradient
            v_grad = beta1 * v_grad + (1-beta1) * grad   # calculating momentum
            s_grad = beta2 * s_grad + (1-beta2) * grad**2   # calculating RMSprop
            v_corr = v_grad / (1-beta1**((pasada+1)*(i+1)))   # bias correction
            s_corr = s_grad / (1-beta2**((pasada+1)*(i+1)))   # bias correction
            theta -= eta * (v_corr / (np.sqrt(s_corr) + eps)).reshape(n,1)   # update theta
            
            Theta = np.hstack((Theta, theta.reshape(n,1)))
        
        eta = eta_ini / (1+10*pasada)  # adjusting the learning rate

    xx = np.linspace(0, pasadas*n_batch, pasadas*n_batch+1)
    for i in range(n):
        plt.plot(xx, Theta[i,:], label = 'theta_%s'%i)
    plt.legend(loc=1)
    plt.title("Minibatch GradientDescent with Adam optimizer and decreasing learning rate")
    plt.xlabel('iterations')
    
    return theta
```

```python
mini_batch_gradientdescent(x, y, theta_ini, 5, 1)
```

In 1 epoch we do 100.000/16 = 6250 iterations. We can see that in the third epoch the values of theta more or less stabilize.  
In this graph you can really see the effect of decreasing the learning rate: after 2 epochs (12.500 iterations) the curve suddenly gets a lot less wiggly.
